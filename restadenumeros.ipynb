## suma de dos numeros

# Solicitamos el primer número 

```python
num1 = int(input("Introduce el primer número: "))
```

# Solicitamos el segundo número

```python
num2 = int(input("Introduce el segundo número: "))
```

# Calcular

```python
suma = num1 + num2
```

# resultado

```python
print("La suma de", num1, "y", num2, "es:", suma)
```
